```python
#Data preprocessing 
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame([[np.nan,np.nan,np.nan],[np.nan,3,4],[np.nan,4,np.nan]])
```

```python
df
```

```python
#check if there is null or not
print(df.isnull())
```

```python
#to drop rows which are null
print(df.dropna())
```

```python
#drop whole column with na values
print(df.dropna(axis=1, how="all"))
```

```python
#fill na with zeros
print(df.fillna(0))
```
